```python
bc2.drop(['품목중분류명'],axis=1,inplace=True)
```

```python
bc2
```

```python
bc2.성별=bc2.성별.apply(lambda x: 0 if x=='남성' else 1) #남성을 0 여성을 1
bc2
```

```python
bc2.연령.unique()
```

```python
bc2.연령=bc2.연령.apply(lambda x:  '20대' if (x == '20대') | (x=='20세 미만') else '30대')

bc2.groupby('연령')['성별'].count()
```

```python
bc2
```

```python
from sklearn.preprocessing import LabelEncoder


le=LabelEncoder()

bc2.품목대분류명=le.fit_transform(bc2.품목대분류명)
bc2.가구생애주기=le.fit_transform(bc2.가구생애주기)
```

```python
bc2.columns
```

```python
bc2=bc2[['기준년월', '품목대분류명', '성별','가구생애주기', '매출금액', '매출건수', '건당_매출금액','연령']]
```

```python
x=bc2.iloc[:,:-1]
y=bc2.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,stratify=y,random_state=1)

print(x_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier

model=RandomForestClassifier(n_estimators=5,random_state=1)

model.fit(x_train,y_train)

train_result=model.predict(x_train)
test_result=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(y_train,train_result))
print(accuracy_score(y_test,test_result))
```

```python
model.feature_importances_
```

```python
import numpy as np

n_features = x.shape[1]
plt.barh(range(n_features),model.feature_importances_,align='center')
plt.yticks(np.arange(n_features),x.columns)
plt.xlabel('중요도')
plt.ylabel('특성')
plt.ylim(-1,n_features)
plt.show()
```

가구생애주기/매출금액, 건당_매출금액 순으로 중요도가 나타남. 다만 모델에 과적합이 발생

```python
from sklearn.ensemble import RandomForestClassifier

model2=RandomForestClassifier(n_estimators=200,max_features='sqrt',random_state=1)

model2.fit(x_train,y_train)

train2_result=model2.predict(x_train)
test2_result=model2.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score


print(accuracy_score(y_train,train2_result))
print(accuracy_score(y_test,test2_result))
```
