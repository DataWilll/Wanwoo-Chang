```python
sh1=shin.drop(['지역구','법정동코드','법정동','집계구코드'],axis=1)
```

```python
sh1.나이.unique()
sh2=sh1.loc[(sh1['나이']==2)|(sh1['나이']==3),]
```

```python
sh3=sh2[['기준년월', '성별', '직장인여부', '급여입금', '가맹점매출입금', '연금입금', '총소비금액','총수신금액', '예적금금액', '신탁금액', '수익증권금액', '신용대출금액', '담보대출금액', '주택대출금액','전세자금대출금액','나이']]
```

```python
sh3
```

```python
x=sh3.iloc[:,:-1]

y=sh3.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1,stratify=y)

print(x_train.shape)
print(x_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier

model=RandomForestClassifier(random_state=1)
```

```python
model.fit(x_train,y_train)
```

```python
train_result=model.predict(x_train)
test_result=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(y_train,train_result))
print(accuracy_score(y_test,test_result))
```

```python
import numpy as np

np.set_printoptions(precision=6, suppress=True)


model.feature_importances_
```
