```python
import pandas as pd

df=pd.read_csv('d:\\data2\\KB국민\\KB_card.csv')
df
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()

df.연령대=le.fit_transform(df.연령대)


```

```python
from datetime import datetime
df['날짜']=df['기준년도'].astype(str)+'/'+df['기준년월'].astype(str)  #년도와 월 컬럼 결합

df['날짜']=pd.to_datetime(df['날짜']).dt.to_period(freq='M') #년-월 형태로만 출력되도록 변경
```

```python
df2=df.drop(['기준년도','기준년월'],axis=1) #기존의 년도와 월 컬럼 삭제
```

```python
df2
```

```python
df2.columns

df3=df2[['날짜','마케팅업종중분류명', '마케팅업종세분류명','성별구분', '연령대', '신용카드보유여부', '체크카드보유여부', '회원수', '매출건수']]
```

```python
df3.연령대.unique()
```

```python
a0=df.loc[df.연령대==0,['날짜','매출건수']]
a1=df.loc[df.연령대==1,['날짜','매출건수']]
a2=df.loc[df.연령대==2,['날짜','매출건수']]
a3=df.loc[df.연령대==3,['날짜','매출건수']]
a4=df.loc[df.연령대==4,['날짜','매출건수']]

```

```python
y0=a0.groupby('날짜')['매출건수'].sum()
y1=a1.groupby('날짜')['매출건수'].sum()
y2=a2.groupby('날짜')['매출건수'].sum()
y3=a3.groupby('날짜')['매출건수'].sum()
y4=a4.groupby('날짜')['매출건수'].sum()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.font_manager as fonm
plt.rcParams['font.family']='Malgun Gothic'


fig=plt.figure(figsize=(20,7))
fig.set_facecolor('white')#배경색
ax=fig.add_subplot() #프레임 생성



x=np.arange(0,25)
xlab=df['날짜'].unique()

ax.plot(x,y0,label='20대 이하')
ax.plot(x,y1,label='30대')
ax.plot(x,y2,label='40대')
ax.plot(x,y3,label='50대')
ax.plot(x,y4,label='60대')

ax.legend()#범례
plt.xticks(x,xlab)
plt.xlabel('날짜')
plt.ylabel('인당_매출건수')
plt.title('연령대별 매출건수')
plt.show()
```

연령대에 구분없이 증감 추세는 비슷하지만 다른 세대에 비해 20대가 증감이 큼 (소비 민감도가 크다)
